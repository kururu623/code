```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)##important
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date)+ 75 ),#标注点函数 spx.asof(date)+ 75是坐标来的
                xytext=(date, spx.asof(date) + 225),#spx.asof(date) + 225 箭头的坐标
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')#important

# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])#设置刻度的初值值和末值，太小会显示不全
ax.set_ylim([600, 1800])

ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
tips = pd.read_csv('examples/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])# Crosstab 交叉列表取值
party_counts
# Not many 1- and 6-person parties
party_counts = party_counts.loc[:, 2:5]
```

```python
party_counts = party_counts.loc[:, 2:5]
party_counts
```

```python
# Normalize to sum to 1
party_pcts = party_counts.div(party_counts.sum(1), axis=0)#进行规格化，使得各行的和为 1，并生成图表
party_pcts
party_pcts.plot.bar()
```

### Scatter or Point Plots

```python
macro = pd.read_csv('examples/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
data[-5:]
data[-5:].diff()
trans_data = np.log(data).diff().dropna()#%log 返回自然对数 diff()是将原来的数据减去移动后的数据,这里是上下相减，dropna删去nan行
trans_data[-5:]#最后五个
```

```python
plt.figure()
```

```python
sns.regplot('m1', 'unemp', data=trans_data)#绘图数据和线性回归模型拟合
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})#diag_kind对角用图
#pairplot:pair是成对的意思，即是说这个用来展现变量两两之间的关系，线性、非线性、相关
```

### Facet Grids and Categorical Data

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])#用于绘制两维变量的关系图，也就是共用x坐标，按照hue分组，画出不同hue种类的y 的值
#参数kind：point默认，bar柱形图，count频次，box箱体，violin提琴，strip散点，swarm分散点
```

```python
sns.factorplot(x='day', y='tip_pct', row='time',
               col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x='tip_pct', y='day', kind='box',
               data=tips[tips.tip_pct < 0.5])
```
